```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
#loading the data set the index to date column
# data of freecodecamp page views (fcc)
fcc_df=pd.read_csv("fcc-forum-pageviews.csv",index_col="date")
fcc_df.info()
```

```python
fcc_df.head()
```

```python
#filter the top 2.5% andd the bottom 2.5%
fcc_df=fcc_df[(fcc_df["value"]>= fcc_df["value"].quantile(0.025)) &
(fcc_df["value"] <= fcc_df["value"].quantile(0.975))]
fcc_df.info()
```

```python
# create a function to draw a line chart 
def draw_line_chart():
    dates=fcc_df.index
    page_views=fcc_df["value"]
    plt.figure(figsize=(18,6))
    plt.plot(dates,page_views,color="r")
    plt.xlabel("Date")
    plt.ylabel("Page Views")
    divisions=8
    x_axis=[i*1000 for i in range(1,divisions+2)]
    x_labels=[fcc_df.index[i] for i in range(0,len(fcc_df.index),len(fcc_df.index)//divisions)]
    plt.xticks(x_axis,x_labels,rotation=60)
    plt.title("Daily freecodecamp page views")
    plt.show()
draw_line_chart()
```

```python
# convert date from str to timestamp and 
#add columns for year,month
fcc_df.index=pd.to_datetime(fcc_df.index)
fcc_df["year"]=fcc_df.index.year
fcc_df["month"]=fcc_df.index.month
fcc_df.head()
```

```python
def draw_bar_chart():
    fcc_groupby_years=fcc_df.groupby(["year","month"],as_index=False)["value"].mean()
    df=pd.DataFrame(pd.DataFrame(fcc_groupby_years).groupby(["year"],as_index=False))
    offset=0
    scaler=1
    width=0.1
    colors=['blue','red','yellow','green','orange','brown','purple','black','silver','pink','indigo','gold']
    for year in df[1]:
        plt.bar((np.arange(len(list(year["month"].index))))+offset,year["value"],0.5,color=[colors[month-1] for month in year['month']] )
        offset+=14
    
    labels=list(np.sort(fcc_df['month'].unique()))
    handles=[plt.Rectangle((0,0),1,1,color=color) for color in colors]
    plt.xticks(np.arange(len(df[1]))*14,fcc_df['year'].unique())
    plt.legend(handles,labels,title='months',loc='upper left',bbox_to_anchor=(1,1))
    plt.show()
draw_bar_chart()
```

```python
def draw_box_plot():
    df_copy=fcc_df.copy()
    df_copy['month_name']=df_copy.index.month_name()
    months_ordered=df_copy.sort_values(["month","year"])
    fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(21,7))
    sb.boxplot(data=df_copy,x="year",y="value",ax=axes[0])
    axes[0].set_title("year wise page trend")
    axes[0].set_xlabel("year")
    axes[0].set_ylabel("page views")
    sb.boxplot(data=months_ordered,x='month_name',y='value')
    axes[1].set_title("month wise page trend")
    axes[1].set_xlabel("month")
    axes[1].set_ylabel("page view")
    plt.xticks(rotation=60)
    
draw_box_plot()
```
