```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
%matplotlib inline
```

```python
sea_levels=pd.read_csv("epa-sea-level.csv")
sea_levels.describe()
```

```python
general_regress=linregress(sea_levels['Year'],sea_levels['CSIRO Adjusted Sea Level'])
from_2000_regress=linregress(sea_levels['Year'][sea_levels["Year"]>=2000],
                             sea_levels['CSIRO Adjusted Sea Level'][sea_levels['Year']>=2000])
def predict_level(regress,x):
    return regress.intercept + regress.slope*x
```

```python
plt.scatter(x=sea_levels['Year'],y=sea_levels['CSIRO Adjusted Sea Level'],label='original data')
extended_years=np.arange(sea_levels['Year'].min(),2051)
from_2000=np.arange(2000,2051)
plt.plot(extended_years,predict_level(general_regress,extended_years),'r',label="fitted line")
plt.plot(from_2000,predict_level(from_2000_regress,from_2000),'g',label="fit from 2000")
plt.xlabel("year")
plt.ylabel("sea level in inches")
plt.title("sea level predictor")
plt.legend()
```
