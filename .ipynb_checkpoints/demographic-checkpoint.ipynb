```python
#filter out high education for males and females
high_edu=["Bachelors","Masters","Doctorate"]
males_high=df.loc[(df["sex"]=="Male")& (df["education"].isin(high_edu))]
females_high=df.loc[(df["sex"]=="Female")&(df["education"].isin(high_edu))]
```

```python
#filter the median of salaries for males and females with bachelors , masters ,doctorate
bachelor_male_median=males_high.loc[males_high["education"]=="Bachelors","hours-per-week"].median()

master_male_median=males_high.loc[males_high["education"]=="Masters","hours-per-week"].median()

doctorate_male_median=males_high.loc[males_high["education"]=="Doctorate","hours-per-week"].median()

bachelor_female_median=females_high.loc[females_high["education"]=="Bachelors","hours-per-week"].median()

master_female_median=females_high.loc[females_high["education"]=="Masters","hours-per-week"].median()

doctorate_female_median=females_high.loc[females_high["education"]=="Doctorate","hours-per-week"].median()
```

```python
# see the result
male_high_medians=[bachelor_male_median,master_male_median,doctorate_male_median]
female_high_medians=[bachelor_female_median,master_female_median,doctorate_female_median]
X=["bachelors","masters","doctorates"]
X_axis=np.arange(len(X))
plt.bar(X_axis-0.2,female_high_medians,0.4,label="Females")
plt.bar(X_axis+0.2,male_high_medians,0.4,label="Males")
plt.xticks(X_axis,X)
plt.xlabel("males and females high  education")
plt.ylabel("the medians for both genders with high education")
plt.legend()
```

```python
# check the distribution of female age 
female_age=pd.DataFrame(df.loc[df["sex"]=="Female","age"]).value_counts()
female_age.sort_index(inplace=True)
plt.hist(x=female_age.tolist(),label="female age distribution")
plt.xlim(female_age.index.min()[0],female_age.index.max()[0])
plt.legend()
```

```python
plt.title("female age scatred")
plt.scatter(female_age.index.tolist(),[x for x in female_age],color="red")
```

```python
# check the capital gain for each race by pie chart
races=np.sort(df["race"].unique())
plt.pie([df.loc[df["race"]==race,"capital-gain"].sum() for race in races],labels=races,autopct="%1.0f%%")
plt.title("capital gain for races")
plt.legend(races,loc="upper left",bbox_to_anchor=(1,1))
```

```python
# check the numbers of each occupation
occu=df["occupation"].value_counts()
plt.bar(np.arange(len(occu)),[oc for oc in occu],label="work occupation")
plt.xticks(np.arange(len(occu)),occu.index,rotation=60)
plt.legend()

```

```python
# check the propotion of different states of people in families
marital_status=df["marital-status"].value_counts()
plt.pie(marital_status,autopct="%1.0f%%")
plt.title("distribution of marital status")
plt.legend(marital_status.index,loc="upper left",bbox_to_anchor=(1,1))
```
